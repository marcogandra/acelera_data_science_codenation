### Desafio Semana 3 - Conhecendo sua base de consumidores: qual estado possui a melhor pontuação de crédito?

```python
import pandas as pd
```

```python
# leitura dos dados
dados = pd.read_csv('desafio1.csv')
dados.head()
```

```python
# verifica o tipo a coluna de pontuação
dados['pontuacao_credito'].dtype
```

```python
# modifica o tipo para float
dados['pontuacao_credito'] = dados['pontuacao_credito'].astype(float)
dados['pontuacao_credito'].dtype
```

```python
# calcula média por estado
media = dados[['estado_residencia','pontuacao_credito']].groupby('estado_residencia').mean()
media
```

```python
# calcula mediana por estado
mediana = dados[['estado_residencia','pontuacao_credito']].groupby('estado_residencia').median()
mediana
```

```python
# calcula moda por estado
moda = dados[['estado_residencia','pontuacao_credito']].groupby('estado_residencia').agg(lambda x : x.value_counts().index[0])
moda
```

```python
# calcula desvio padrão por estado
desvio = dados[['estado_residencia','pontuacao_credito']].groupby('estado_residencia').std()
desvio
```

```python
# cria dicionário
dic = {'PR': {'media': media['pontuacao_credito'][0],
              'mediana': mediana['pontuacao_credito'][0],
              'moda': moda['pontuacao_credito'][0],
              'desvio_padrao': desvio['pontuacao_credito'][0]},
       'RS': {'media': media['pontuacao_credito'][1],
              'mediana': mediana['pontuacao_credito'][1],
              'moda': moda['pontuacao_credito'][1],
              'desvio_padrao': desvio['pontuacao_credito'][1]},
       'SC': {'media': media['pontuacao_credito'][2],
              'mediana': mediana['pontuacao_credito'][2],
              'moda': moda['pontuacao_credito'][2],
              'desvio_padrao': desvio['pontuacao_credito'][2]}}
dic
```

```python
# salva arquivo json
import json

with open('submission.json', 'w') as fp:
    json.dump(dic, fp)
```
